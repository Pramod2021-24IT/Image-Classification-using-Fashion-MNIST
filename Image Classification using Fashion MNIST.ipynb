# Data Normalization

### We then normalize the data dimensions so that they are approximately the same scale.

```python
X_train_n = X_train_full/255.0
X_test_n = X_test/255.0
```

## Split the data into train/validation/test datasets
In the earlier step of importing the data, we had 60,000 datasets for training and 10,000 test datasets. Now we further split the train/validation. Here is how each type of dataset is used in deep learning:

* Training Data- used for training the model
* Validation data- used for tuning the hyperparameters and evaluating the models
* Test data- used to test the model after the model has gone through initial vetting by the validation set.

```python
X_valid, X_train = X_train_n[:5000], X_train_n[5000:]
y_valid, y_train = y_train_full[:5000], y_train_full[5000:]
X_test = X_test_n
```

- Training Data(55,000)- X_train and y_train - for trainig the model

## Building the Neural Network using Keras

```python
np.random.seed(42)
tf.random.set_seed(42)
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Flatten(input_shape=[28, 28]))
model.add(tf.keras.layers.Dense(300, activation="relu"))
model.add(tf.keras.layers.Dense(100, activation="relu"))
model.add(tf.keras.layers.Dense(10, activation="softmax"))
```
